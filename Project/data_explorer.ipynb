```python
import pandas as pd
import json
import os
import numpy as np
```

```python
data_folder= 'data/'

df = None
for file in os.listdir(data_folder):
    file_path = os.path.join(data_folder, file)
    with open(file_path) as sd:
        data = json.load(sd)
    cdf = pd.DataFrame(data)
    cdf['cuisine'] = file.split(".")[0]
    df = cdf if df is None else pd.concat([df, cdf])

df = df[['flavors', 'id', 'ingredients', 'recipeName', 'cuisine']]
# print(len(df[df['flavors'] == np.nan]))#
print(len(df[df['flavors'].apply(lambda x: x is None)])*1./len(df))
df.head(-5)
```
