```python
import pandas as pd
import json
import os
import numpy as np
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_folder= 'data/'

df = None
for file in os.listdir(data_folder):
    if 'DS_' in file:
        continue
    file_path = os.path.join(data_folder, file)
    with open(file_path) as sd:
        data = json.load(sd)['matches']
    cdf = pd.DataFrame(data)
    cdf['cuisine'] = file.split(".")[0].split("_")[-2]
    df = cdf if df is None else pd.concat([df, cdf])
    print("For {} cuisine we have {} recipes (from {} recipes)".format(file.split(".")[0], len(cdf[cdf['flavors'].apply(lambda x: not x is None)]), len(cdf)))
    
df = df[['flavors', 'id', 'ingredients', 'recipeName', 'cuisine']]

withnone = (len(df[df['flavors'].apply(lambda x: x is None)]))
print("In total we have {} recipes from which {} do not have flavors.".format(len(df), withnone))
print("Only {} recipes have flavours.".format(len(df)-withnone))
```

```python
#Scaling
rank = {0:0,0.16666667 :1,0.33333333:2, 0.5:3,0.66666667:4,0.83333333:5,1:6}
```

```python
df['flavors'].iloc[0]
df['bitter'] = None
df['salty'] = None
df['piquant']= None
df['sour']= None
df['sweet']= None

def separate_flav_column(x):
    if x.flavors is not None:
        for (k,v)in  x.flavors.items():
            if(v is not None):
                x[k]= rank[round(v,8)]
            else:
                x[k] = v
    return x
df_2 = df.apply(separate_flav_column,axis=1)
```

```python
groupped_means_flavors = df_2.groupby('cuisine').agg('mean')
groupped_stds_flavors = df_2.groupby('cuisine').agg('std')
```

```python
display(HTML('<h3>Mean flavors</h3>'))
display(groupped_means_flavors)
display(HTML('<h3>STDs of  flavors</h3>'))
display(groupped_stds_flavors)
```

```python
scaled_groupped_means_flavors = groupped_means_flavors.apply(lambda x:(x - x.min())/(x.max()-x.min()))
scaled_groupped_stds_flavors = groupped_stds_flavors.apply(lambda x:(x - x.min())/(x.max()-x.min()))
```

```python
display(HTML('<h3>Scaled Mean flavors</h3>'))
display(scaled_groupped_means_flavors)
```

```python
display(HTML('<h3>Covariance of flavours on the whole dataset</h3>'))
display(df_2.loc[:,['bitter','sour','salty','meaty','piquant','sweet']].corr())

for i in df_2.cuisine.unique().tolist():
    display(HTML('<h3>Covariance of flavours in {} cuisine</h3>'.format(i)))
    display(df_2.loc[df_2.cuisine == i,['bitter','sour','salty','meaty','piquant','sweet']].corr())
```

```python
colors = ['#FF6633', '#FFB399', '#FF33FF', '#FFFF99', '#00B3E6', 
		  '#E6B333', '#3366E6', '#999966', '#99FF99', '#B34D4D',
		  '#80B300', '#809900', '#E6B3B3', '#6680B3', '#66991A', 
		  '#FF99E6', '#CCFF1A', '#FF1A66', '#E6331A', '#33FFCC',
		  '#66994D', '#B366CC', '#4D8000', '#B33300', '#CC80CC', 
		  '#66664D', '#991AFF', '#E666FF', '#4DB3FF', '#1AB399',
		  '#E666B3', '#33991A', '#CC9999', '#B3B31A', '#00E680', 
		  '#4D8066', '#809980', '#E6FF80', '#1AFF33', '#999933',
		  '#FF3380', '#CCCC00', '#66E64D', '#4D80CC', '#9900B3', 
		  '#E64D66', '#4DB380', '#FF4D4D', '#99E6E6', '#6666FF'][:19]
def plot_features(x,colors):
    cuisines = x.cuisine.unique()
    patches = [mpatches.Patch(color=c, label= l) for (c,l) in zip(colors,cuisines)]
    f, axarr = plt.subplots(4,2)
    f.set_size_inches(18.5, 10.5)
    plt.legend( handles=patches,loc="upper left", bbox_to_anchor=[0, 1],ncol=2, shadow=True, title="Legend", fancybox=True)
    for i,flavor in enumerate(['bitter','sour','sweet','salty','meaty','piquant']):
        vects = [df_2.loc[df_2.cuisine == c,flavor].dropna() for c in cuisines]
        axarr[int(i/2.0),i%2].set_title(flavor)
        axarr[int(i/2.0),i%2].hist(vects,stacked = True,color = colors,bins=40,normed = True)
        
        
plot_features(df_2,colors)


# ADD box plots
```
