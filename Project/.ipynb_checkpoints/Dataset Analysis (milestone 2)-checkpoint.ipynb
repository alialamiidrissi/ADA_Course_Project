## Introduction

Diets and food habits may vary widely from country to country in terms of ingredients and cooking techniques. Nevertheless, dishes from different regions may share similarities in flavours and tastes. This project aims to help people discover new recipes matching their preferences. Our analysis will be based on Yummly, a platform mainly used in North America containing recipes from different countries. First, we will characterise the taste preferences in a group of countries along with the nutritional intake of an average dish. Second, we will analyse similarities between cuisines in terms of flavours with the purpose of building a recipe recommender that matches one'|s taste.


**Data collection**

Using the Yummly API, we collect a number of recipes (see `spider.py` for the scrapper code) for 19 cuisines along with their ingredients. Our current analysis focuses only on the main course dishes.

The dataset can be found under the folder `data` where each json file correpond to the list of recipes of a cuisine.

**Research questions taken into account**

We will try to answer the following question in our exploratory analysis:
- What are the most representative ingredients of a cuisine?
- What are the dominant flavours of a cuisine?

## Helper functions

```python
def load_data(data_folder='data/'):
    '''Helper function to load and concat data'''
    df = None
    for file in os.listdir(data_folder):
        if 'DS_' in file:
            continue
        file_path = os.path.join(data_folder, file)
        with open(file_path) as sd:
            data = json.load(sd)['matches']
        cdf = pd.DataFrame(data)
        cdf['cuisine'] = file.split(".")[0].split("_")[-2]
        df = cdf if df is None else pd.concat([df, cdf])
        #print("For {} cuisine we have {} recipes (from {} recipes)".format(file.split(".")[0], len(cdf[cdf['flavors'].apply(lambda x: not x is None)]), len(cdf)))
    df = df[['flavors', 'id', 'ingredients', 'recipeName', 'cuisine']]
    return df

def draw_radar_graph(df, index,
                    title='', ax=None, ylim=None):
    ''' Helper function to draw radar function from a DataFrame '''
    labels = np.array(['piquant', 'meaty', 'bitter', 'sweet', 'sour', 'salty'])
    stats = df.loc[index, labels].values
    angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
    stats=np.concatenate((stats,[stats[0]])) 
    angles=np.concatenate((angles,[angles[0]]))
    if ax is None:
        fig= plt.figure()
        ax = fig.add_subplot(111, polar=True)
    ax.plot(angles, stats, 'o-', linewidth=2)
    ax.fill(angles, stats, alpha=0.25)
    ax.set_thetagrids(angles * 180/np.pi, labels)
    ax.set_title(title)
    if not ylim is None:
        ax.set_ylim((0,ylim))
    return ax

def draw_distribution_ingredients(df, ax=None):
    ''' Helper function to draw the distribution of ingredients in a data frame'''
    get_list_ingredients(df.loc[df.cuisine == cuisine]).sort_values(by='Number of apparition', ascending=False).head(20).plot(x='Ingredient', y='Number of apparition', kind='barh', ax = ax )
    ax.set_title('Distribution of the '+cuisine+' cuisine ingredients',fontsize= 15)
    ax.tick_params(direction='out', length=6, width=4)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(15)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(15)
    ax.set_ylabel('Ingredients', fontsize=16)
    return ax
    
def get_list_ingredients(df):
    ''' Helper function to get the list of ingredients from a DataFrame'''
    possible_ingredients = {}
    for a  in df['ingredients']:
        for ing in a:
            possible_ingredients[ing] = 1 if(not ing in possible_ingredients) else possible_ingredients[ing]+1
    ing_data = np.array([(key,possible_ingredients[key]) for key in possible_ingredients.keys()])
    df_ing = pd.DataFrame(data={'Ingredient': ing_data[:,0], 'Number of apparition':ing_data[:,1].astype(np.int32)})
    return df_ing
```

### Filtering the recipes

To improve our result, we filter the recipes by keeping only the recipes with ingredients widely used in a given cuisine. To do so, we compute the TFIDF scores for every ingredient and input the resulting features to a multi-class logistical regression model. This model will try to match the recipes to the closest cuisines.
We keep a recipe if and only if it was classified correctly by our Logistical regression with a probability higher a threshold.

```python
# Create a set containing all ingredients
ingr = set()
df.ingredients.apply(lambda x: [ingr.add(y) for y in x])

# Converting ingredients to text
text_ingredients = df.ingredients.apply(lambda l_ingredients : functools.reduce(lambda acc,x : acc +' ' +x,map(lambda ingred: ingred.translate(str.maketrans('', '', string.whitespace)),l_ingredients)))
```

```python
# Vectorize recipes
vectorizer = TfidfVectorizer(stop_words=None)
vectors = vectorizer.fit_transform(text_ingredients)
```

```python
df.set_index
cuisines_label = pd.get_dummies(df.cuisine)
cuisines_classes = cuisines_label.columns.tolist()

# Convert cuisines to categories
df.cuisine = df.cuisine.astype('category')

# Set the logistical regression model parameters and train it
logistic = LogisticRegression(multi_class= 'multinomial',solver='lbfgs')
logistic.fit(vectors,df.cuisine.astype('category').cat.codes)
predicted_probas = logistic.predict_proba(vectors)
```

```python
print('Model accuracy:',logistic.score(vectors,df.cuisine.astype('category').cat.codes))
```

```python
#Create a dataframe containing the predicted classes along with the probability of every class
recipes_to_cuisine_1 = pd.DataFrame(predicted_probas,columns =  df.cuisine.cat.categories)
recipes_to_cuisine_1 = pd.concat([df.id,recipes_to_cuisine_1],axis = 1)
recipes_to_cuisine_1 = recipes_to_cuisine_1.set_index('id',drop = True)
recipes_to_cuisine_1['classes'] = logistic.predict(vectors)
```

```python
def apply_func(x):
    temp =recipes_to_cuisine_1.loc[x.id,'classes'].astype(int)
    return (recipes_to_cuisine_1.loc[x.id,x.cuisine] >= 0.6) and (df.cuisine.cat.categories.tolist()[temp] == x.cuisine)
# Add a column in the original dataframe that determines which entries will be kept
df['keep'] = df.apply(apply_func  ,axis = 1)
df_filtered = df[df['keep']]
```

```python
def draw_distribution_flavours(df, ax=None):
    ''' Helper function to draw the distribution of flavours in a data frame'''
    df.loc[df.cuisine==cuisine, ['piquant', 'meaty', 'bitter', 'sweet', 'sour', 'salty']].plot(kind='box', ax=ax)
    ax.set_title(cuisine)
    ax.set_ylim(ax.get_ylim()[0], 6)

# Draw the distribution of flavours for every cuisine
f,ax = plt.subplots(5,4, figsize=(18,18))
f.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.3)
for i, cuisine in enumerate(df_filtered['cuisine'].unique()):
    draw_distribution_flavours(df_filtered[df.cuisine==cuisine], ax=ax[i//4,i%4])
d=ax[-1, -1].axis('off')
```

```python
flavors_cuisines = df[df['keep']].groupby(by='cuisine').agg('mean')
display(flavors_cuisines)
flavors_cuisines[['piquant', 'meaty', 'bitter', 'sweet', 'sour', 'salty']] = flavors_cuisines[['piquant', 'meaty', 'bitter', 'sweet', 'sour', 'salty']].apply(lambda x: (x-x.min())/(x.max()-x.min()))
    
# Plot radar charts
f,ax = plt.subplots(5,4, figsize=(18,18), subplot_kw=dict(polar=True))
f.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.3)
for i, cuisine in enumerate(flavors_cuisines.index):
    draw_radar_graph(flavors_cuisines, cuisine, title=cuisine, ax=ax[i//4,i%4], ylim=1)
d=ax[-1, -1].axis('off')
```
